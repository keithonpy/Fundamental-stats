## Motivation of Neural Network

Similar to Classification, classifer uses a mapping function (mapping for $\phi(x)$)
$$
\hat{y} = sign(\theta \cdot \phi(x))
$$
Neural network passes the linear function input to non-linear function called activation function

## Activation function of neural network

1. Linear function: 
$$
f(Z) = Z
$$
1. ReLU (Rectified linear unit): 
$$
f(Z) = \begin{cases}
Z & \text{if} Z > 0 \\
0 & \text{if} Z \leq 0
\end{cases}
$$
1. Hyperbolic tangent function: 
$$
f(Z) = \frac{e^{Z} - e^{-Z}}{e^{Z} + e^{-Z}}
$$
